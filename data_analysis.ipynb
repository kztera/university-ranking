- Tập dữ liệu được lấy từ `timeshighereducation.com` với các thông tin về xếp hạng các trường đại học trên toàn thế giới năm 2023
- Tập dữ liệu gồm 1799 dòng và 13 cột gồm:
  - rank: xếp hạng của trường đại học
  - name: tên của trường đại học
  - country: quốc gia của trường đại học
  - number_of_students: số lượng sinh viên
  - student_per_staff: số lượng sinh viên trên mỗi giáo viên
  - international_students: phần trăm sinh viên quốc tế
  - female_male_ratio: tỉ lệ sinh viên nữ / nam
  - overall_score: điểm tổng thể
  - teaching_score: điểm giảng dạy
  - research_score: điểm nghiên cứu
  - citations_score: điểm trích dẫn
  - industry_income_score: điểm thu nhập từ ngành công nghiệp
  - international_outlook_score: điểm quan điểm quốc tế

Các điểm đánh giá được giải thích tại [đây](https://www.timeshighereducation.com/student/advice/world-university-rankings-explained).

````python
county_count = df['country'].value_counts().to_frame("count")

county_count = county_count.rename_axis('country').reset_index()

county_count = county_count.reset_index(drop=True)

county_count
````

````python
# export to excel without column index

county_count.to_excel('data/country_count_2023.xlsx', index=False)
````
